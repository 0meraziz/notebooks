### Data Preprocessing

#### Missing Names

```python
missing_names = astronaut_data['original_name'].isnull() #Creating a boolean mask to filter the DataFrame
astronaut_data[missing_names][['name', 'original_name', 'nationality', 'occupation', 'mission_title']]  #Viewing only a subset of the DataFrame
```

The `original_name` column is the name of the astronaut in their original language. This information will most likely be irrelevant to our analysis so we can either drop the column entirely or ignore the missing values. In this case, we will drop the column

```python
astronaut_data.drop(columns='original_name', inplace=True)
```

#### Selection 

```python
missing_selection = astronaut_data['selection'].isnull() #Creating a boolean mask to filter the DataFrame
astronaut_data[missing_selection][['name', 'sex', 'nationality', 'selection', 'occupation']] #Viewing only a subset of the DataFrame
```

The observations with missing values for `selection` all have the `occupation` 'space tourist'. This suggests that `NaN` in this case means 'did not attend a selection program'

We can also note that the `occupation` 'space tourist' has different entries. This is also something we can investigate

```python
space_tourists = astronaut_data['occupation'].str.contains('tourist') #Filter rows where the occupation contains 'tourist'
astronaut_data[space_tourists][['name', 'sex', 'nationality', 'selection', 'occupation']]
```

**Notes**

- The occupation 'space tourist' has three different entries: Other(Space tourist), Other(space tourist) and Space tourist.
- There is inconsistency in the `selection` category, where both 'Space Tourist' and `NaN` have been used to specify no selection program was attended

For consistency, we will set both `occupation` and `selection` to 'Space Tourist' 

```python
#Using pd.apply to mutate the entries 
astronaut_data['occupation'] = \
astronaut_data['occupation'].apply(lambda x: 'Space Tourist' if 'tourist' in x else x) 

#Replacing NaN values in selection with Space Tourist
astronaut_data['selection'].fillna('Space Tourist', inplace=True) 

#Replacing 'Space Tourist 2' with 'Space Tourist'
astronaut_data['selection'].replace('Space Tourist 2', 'Space Tourist', inplace=True) 
```

We can now check whether our changes have taken place

```python
astronaut_data[space_tourists][['name', 'sex', 'nationality', 'selection', 'occupation']]
```

#### Mission Title/ Ascend/Decend Shuttle

```python
missing_title = astronaut_data['mission_title'].isnull()
astronaut_data[missing_title]
```

```python
missing_shuttle = astronaut_data['ascend_shuttle'].isnull()
astronaut_data[missing_shuttle]
```

**Notes**

- There are two observations with missing values, one with a missing mission title and one with missing values for ascent and descent shuttles
- Since these are only two observations, we can either try and fill in this missing information or we could just ignore/remove these rows

#### Adding Columns

Since we have both `year_of_birth` and `year_of_mission`, we can add a new `age_at_mission` column

```python
astronaut_data['age_at_mission'] = astronaut_data['year_of_mission'] - astronaut_data['year_of_birth']
```

We can also add a `decade` column to monitor changes over the years

```python
#By subtracting the year in the ones place, we can obtain the decade i.e. 1977 - 7 = 1970
astronaut_data['decade'] = astronaut_data['year_of_mission'] - astronaut_data['year_of_mission']%10
```

### Data Preprocessing

#### Cost

```python
#Removing the whitespace in the column name
mission_data.rename(columns={' Cost': 'Cost'}, inplace=True)
```

```python
percent_missing = mission_data.Cost.isnull().sum() * 100 /len(mission_data.Cost)
print(f'{percent_missing:.2f}% values missing')
```

Over 77% of our values for `Cost` are missing, this would normally mean that there is insuficient information in this column for any meaningful analysis. However, since there are still almost 1000 values, maybe we have complete information for some specific companies

```python
#Filter rows that don't have cost
cost_missing = mission_data['Cost'].isnull()

companies_cost_missing = mission_data[cost_missing]['Company Name'].unique() #List of companies with incomplete cost information
companies_cost_all = [company for company in mission_data['Company Name'].unique() if company not in companies_cost_missing] #List of compamies with complete cost information

print(f'Complete cost information on {len(companies_cost_all)} companies')
```

```python
mission_data[mission_data['Company Name'].isin(companies_cost_all)].sort_values('Company Name')
```

In the table above you might note that the costs are the same for all missions by the same company. 

We can take Eurockot as an example:

```python
eurockot = mission_data['Company Name'] == 'Eurockot'
mission_data[eurockot][['Company Name', 'Date', 'Detail', 'Cost']]
```

Even though details of the rockets have changed and there are 18 years between the first and last missions carried out by Eurockot, the cost of a mission has stayed the same at $41.8m. This coupled with the fact that the values for cost are largely missing likely suggests that this column should be treated with caution and therefore, for the purposes of this analysis, we will ignore it.

#### Adding Columns

We can identify the country of the launch from the `Location`

```python
mission_data['Country'] = [x[-1].strip() for x in mission_data['Location'].str.split(',')]
```

We can also add a `Mission Year` Column

```python
mission_data['Mission Year'] = pd.to_datetime(mission_data['Date'], format='%a %b %d, %Y', exact=False).dt.year
```

## Astronauts 

### General information

Here is some general information we can obtain from this data:

```python
total_space_time = astronaut_data['hours_mission'].sum()
oldest_at_mission = astronaut_data[['name','age_at_mission', 'year_of_mission']].nlargest(1, 'age_at_mission')
youngest_at_mission = astronaut_data[['name','age_at_mission', 'year_of_mission']].nsmallest(1, 'age_at_mission')
distinct_nationalities = astronaut_data['nationality'].nunique()
prop_female = astronaut_data['sex'].value_counts(normalize=True)

print(f"Did you know?\n")
print(f"> Humans have spent {total_space_time:.0f} manhours in space. That's {total_space_time/(24*365):.0f} years!")
print(f"> The oldest astronaut to go into space is {oldest_at_mission.iloc[0,0]}. They were {oldest_at_mission.iloc[0,1]} years at the time!")
print(f"> The youngest astronaut is {youngest_at_mission.iloc[0,0]}. They were the first astronaut aged {youngest_at_mission.iloc[0,1]}")
print(f"> People from {distinct_nationalities} different nations have ventured into space")
print(f"> Only {prop_female['female']*100:.0f}% of all space travellers have been female")
```

We can also look at the data grouped by sex:

```python
#We can obtain the number of female astronauts each decade through grouping by sex, and counting the unique entries
female_ratio = astronaut_data.groupby(by=['decade', 'sex']).nunique()['number']

#Grouping by decade again and using apply, we can calculate the proportions 
female_ratio = female_ratio.groupby('decade').apply(lambda x  : x *100/ x.sum()).unstack()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12,5)) #Creating a figure with 2 subplots

ax.flatten() 
ax[0].bar(female_ratio.index, 100, width=9.0, label='Men', color=['darkcyan'])
ax[0].bar(female_ratio.index, female_ratio['female'], width=9.0, label='Women', color=['salmon'])

ax[0].set_xlabel('Decade') 
ax[0].set_title('Proportion of Astronauts by Sex')
ax[0].legend()

sns.lineplot(data=female_ratio.replace(np.nan,0)['female'], ax=ax[1], lw=2, color='salmon')
ax[1].set_title('Percentage of Female Astronauts')
ax[1].set_xlabel('Decade')
ax[1].set_ylabel('')

plt.suptitle('Though female reprsentation is still very low (Below 20%), we can see that it is has been getting better with each decade')
plt.show()
```

### Missions

We look at the information relating to each particular mission:

```python
#Grouping by mission title > decade of mission| total mission hours
#Grouping by decade > average mission hours
avg_mission_hours_decade = astronaut_data.groupby('mission_title').agg({'decade':'last',
                                            'hours_mission':'sum'}).groupby('decade').mean()

#Grouping by decade and counting unique missions 
num_missions_decade = astronaut_data.groupby('decade').nunique()['mission_title']
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15,5))

ax.flatten()
sns.lineplot(data=num_missions_decade, ax=ax[0], lw=2, color='royalblue')
ax[0].set_xlabel('Year', size=12)
ax[0].set_ylabel('Number of missions', size=12)
ax[0].set_title('Missions each year')


sns.lineplot(data=avg_mission_hours_decade['hours_mission'], ax=ax[1], lw=2, color='firebrick')
ax[1].set_title('Hours/Mission')
ax[1].set_xlabel('Year', size=12)
ax[1].set_ylabel('')
plt.suptitle('Astronauts are spending more hours per mission than ever before', fontsize=15)
plt.show()
```

From the figures, we can see that even though the number of launches per year has been going down, the number of hours astronauts are spending on missions is increasing. 

## Missions

### Missions by country

```python
mission_data['Failure'] = mission_data['Status Mission'].map(lambda x: 0 if x == 'Success' else 1)
```

```python
country_missions = mission_data.groupby(['Country']).agg({'Company Name': 'count',
                                                    'Failure': 'sum'}).sort_values('Company Name')[5:] #removing the 5 lowest values for the fig
```

```python
fig, ax = plt.subplots(figsize=(9,5))

ax.barh(country_missions.index, country_missions['Company Name'], label='Total Missions', color=['mediumblue'])
ax.barh(country_missions.index, country_missions['Failure'], label='Failed', color=['firebrick'])

ax.set_title('Number of missions by launch location')
ax.legend(loc='lower right')
plt.show();
```

As you would expect, Russia and the U.S. are the two leading countries in terms of missions. It might be interesting to compare the two.

```python
countries= ['Russia', 'USA']
us_rus_filt = mission_data.query('Country in @countries')
us_rus = us_rus_filt.groupby(['Mission Year', 'Country']).count()[['Company Name']].unstack()
```

We can also get the date of the apollo 11 launch from the dataset!

```python
apollo_11 = mission_data[mission_data['Detail'].str.contains('Apollo 11')]

moon_landing_year = apollo_11['Mission Year'].iloc[0]
```

```python
plt.figure(figsize=(9,5))

#Plotting Lines
sns.lineplot(x=us_rus.index, y=us_rus['Company Name', 'Russia'], label='Russia', lw=2, color='firebrick')
sns.lineplot(x=us_rus.index, y=us_rus['Company Name', 'USA'], label='USA', lw=2, color='blue')
plt.ylabel('Number of Launches', size=12)
plt.xlabel('Year', size=12)

#Highlight Region
plt.axvspan(1955, 1990, alpha = 0.5, color = 'bisque')
plt.text(x = 1955, y = 70, s = 'Cold War Period', fontsize=12)

#Marking the moon landing
plt.axvline(x = moon_landing_year, lw = 2, color = 'goldenrod', linestyle = '--')
plt.text(x = moon_landing_year+1, y = 20, s = 'First Moon Landing', rotation = 90, fontsize = 10)

plt.title('The Space Race', size=14)
plt.legend()
plt.show();
```

Interestingly enough, from the above figure we learn that, in the year Apollo 11 landed on the moon, Russia had **double** the launches as compared to the U.S. A reason for this disparity could be that the U.S. was at a more advanced stage in its program while Russia was catching up. 

We can also see that the Russian space program became much less active after the collapse of the Soveit Union.
